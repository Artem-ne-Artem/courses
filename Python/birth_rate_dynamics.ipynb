### Получить данные из HTML таблицы с данными по рождаемости в России.
### Визуализировать на графике.
#### https://worldtable.info/gosudarstvo/tablica-rozhdaemosti-po-godam-rossija.html

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
birth = pd.read_html("https://worldtable.info/gosudarstvo/tablica-rozhdaemosti-po-godam-rossija.html"\
                     ,header=0\
                     ,parse_dates=['Год'])
birth = birth[0]
```

```python
# переименовываем поля
birth = birth.rename (columns = {'Год' : 'year', 'Количество родившихся чел.' : 'qty'})
# чистим числа от пробелов, переводим в int
birth['qty'] = birth['qty'].str.replace(' ','').astype(int)
birth.info()
```

```python
birth.plot(x='year',
           y='qty',
           style='--',
           rot=45,
           figsize=(20, 5),           
           grid=True)
plt.xlabel('Год',fontsize=16, color='black')
plt.ylabel('Количество родившихся чел. (млн)',fontsize=16, color='black')
plt.title('Динамика рождаемости', fontsize=20, color='black');
plt.show()
```
